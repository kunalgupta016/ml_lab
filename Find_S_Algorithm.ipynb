# Find-S Algorithm
Using EnjoySport Dataset

```

import pandas as pd

df = pd.read_csv("data.csv")
df
```

## Separate Features and Target

```

X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
X, y
```

## Find-S Algorithm Implementation

```

hypothesis = ['Ø'] * X.shape[1]

for i in range(len(X)):
    if y[i].lower() == 'yes':
        for j in range(len(hypothesis)):
            if hypothesis[j] == 'Ø':
                hypothesis[j] = X[i][j]
            elif hypothesis[j] != X[i][j]:
                hypothesis[j] = '?'

hypothesis
```
